```python
zone_set1 = [[] for _ in range(Total_Zones)]

for _, row in bb.iterrows():
    zone = row['Zone']
    b_index = row['B_index']
    influence = row['Influence']
    zone_set1[zone].append([influence, b_index])
    
for element in zone_set1:
    sorted(element)

print(zone_set1) 
```

```python
zone_sets = [set() for _ in range(Total_Zones)]

for _, row in bb.iterrows():
    zone = row['Zone']
    b_index = row['B_index']
    zone_sets[zone].add(b_index)

print(zone_sets) 
```

```python
total_advertisers = len(advertiser)
total_advertisers
```

```python
advertiser = advertiser.sort_values(by='Payment_Demand_Ratio', ascending=False)
advertiser
```

```python
advertiser_df = pd.DataFrame(advertiser)
```

```python
advertisers = advertiser_df.to_dict(orient='records')
```

```python
for adv in advertisers:
    adv['allocation'] = [set(),set(),set(),set(),set()]
advertisers
```

```python
ub['Influenced Billboards'] = ub['Influenced Billboards'].apply(lambda x: ast.literal_eval(x))
ub
```

```python
def influence(BS):
    influence_score = 0
    for _, row in ub.iterrows():
        influenced_billboards = row['Influenced Billboards']
        intersection_result = {key: value for key, value in influenced_billboards.items() if key in BS}
        product = 1.0
        for value in intersection_result.values():
            product *= (1 - value)
        influence_score += 1 - product
    
    return influence_score
```

```python
def Regret(total_influence, demand, payment):
    if demand > total_influence:
        penalty_ratio = 0.5
        regret = payment * (1 - (penalty_ratio * (total_influence/demand)))
    
    else:
        regret = payment * ((total_influence - demand)/demand)
    
    return regret
```

```python
def func(x, S, payment, demand):
    a1 = Regret(influence(S), demand, payment)
    a2 = Regret(influence(S.union({x})), demand, payment)
    a = a1-a2
    b = bb[bb['B_index'] == x]['Influence'].values[0]
    return a/b
```

```python
eps = 0.01
excessive_regret = unsatisfied_regret = 0
satisfied_adv=[]
unsatisfied_adv = []
while True:
    for adv in advertisers:
        advertiser_id = adv['Id']
        if advertiser_id in satisfied_adv:
            continue
        zonal_demand = adv['Zonal_Demand']
        zone_set2 = copy.deepcopy(zone_set1)      
        for i, demand in enumerate(zonal_demand):
            temp_set = set()
            temp_influence = 0
    #         print(len(zone_sets[i]))
            while temp_influence < demand and len(zone_set2[i]) > 0:
                temp_set.add(zone_set2[i][0][1])
                temp_influence += zone_set2[i][0][0]
                del zone_set2[i][0]
             
            while demand > influence(adv['allocation'][i]) and len(zone_sets[i]) > 0:
                temp_size = int((len(zone_sets[i]) * math.log(1/eps)) / len(temp_set))
    #             print(len(zone_sets[i]),len(temp_set),temp_size)
                if temp_size == 0:
                    temp_size = len(zone_sets[i])
                if temp_size > len(zone_sets[i]):
                    temp_size = len(zone_sets[i])
                P = random.sample(zone_sets[i], temp_size)
#                 print(P)
                output = max(P, key=lambda x: func(x, adv['allocation'][i], adv['Payment'], adv['Demand']))
#                 print('output',output)
                adv['allocation'][i].add(output)
                zone_sets[i].remove(output)
#                 print('zone_sets',zone_sets)

    count = 0
    satisfied_adv.clear()
    unsatisfied_adv.clear()
    
    for adv in advertisers:
        supply = 0
        for j in range(0, 5):
            supply += influence(adv['allocation'][j])
        demand = adv['Demand']
        if demand < supply:
            satisfied_adv.append(adv['Id'])
        elif demand > supply:
            count += 1
            unsatisfied_adv.append(adv)  # Store the entire advertiser dictionary
    
    print('satisfied_adv', satisfied_adv)
    print('unsatisfied_adv', [adv['Id'] for adv in unsatisfied_adv])

    if count > 1:
        min_demand_advertiser = min(unsatisfied_adv, key=lambda x: x['Payment_Demand_Ratio'])
        print('min_demand_advertiser', min_demand_advertiser['Id'])
        for i in range(len(min_demand_advertiser['allocation'])):
            zone_sets[i] = min_demand_advertiser['allocation'][i]
        advertisers.remove(min_demand_advertiser)
        count -= 1
    else:
        break           
```

```python
excessive_regret = unsatisfied_regret = count = 0
for adv in advertisers:
    supply = 0
    for j in range(0, 3):
        supply += influence(adv['allocation'][j])
    demand = adv['Demand']
    payment = adv['Payment']
    if demand > supply:
        penalty_ratio = 0.5
        unsatisfied_regret += payment * (1 - (penalty_ratio * (supply/demand)))
    else:
        excessive_regret += payment * ((supply - demand)/demand)
        count+=1

print('The number of advertiser satisfied', count, 'out of', total_advertisers)
print('Excessive Regret', excessive_regret)
print('unsatisfied Regret', unsatisfied_regret)
```

```python
total_regret = excessive_regret + unsatisfied_regret
print(total_regret)
end1 = time.time()
print(f"Total runtime of the program is {end1 - begin1}")
```

# Advertiser Driven Starts from Here

```python
def Regret_new(advertisers):
    excessive_regret = unsatisfied_regret = count = 0
    for adv in advertisers:
        supply = 0
        for j in range(0, 5):
            supply += influence(adv['allocation'][j])
        demand = adv['Demand']
        payment = adv['Payment']
        if demand > supply:
            penalty_ratio = 0.5
            unsatisfied_regret += payment * (1 - (penalty_ratio * (supply/demand)))
        else:
            excessive_regret += payment * ((supply - demand)/demand)
    total_regret = unsatisfied_regret + excessive_regret
    return total_regret
```

```python
import copy
def AdvertiserDriven(advertisers):
    excessive_regret = unsatisfied_regret = 0
    advertiser_best = copy.deepcopy(advertisers)
    advertiser_temp = copy.deepcopy(advertisers)
    loop_counter = 0
    
    while True:
        advertiser_can = copy.deepcopy(advertiser_best)
        loop_counter = loop_counter + 1
        for i in range(len(advertiser_can)):
            for j in range(i + 1, len(advertiser_can)):
                adv1 = advertiser_can[i]
                adv2 = advertiser_can[j]
                
                R_old = Regret_new(advertiser_temp)
                print('R_old: ', R_old)
                
                # Swap the 'allocation' values between adv1 and adv2
                adv1['allocation'], adv2['allocation'] = adv2['allocation'], adv1['allocation']
                
                R_new = Regret_new(advertiser_can)
                print('R_new: ', R_new)
                
                if R_new <= R_old:
                    advertiser_temp = copy.deepcopy(advertiser_can)
                    continue   
                else:
                    # Swap allocations back to the original state
                    adv1['allocation'], adv2['allocation'] = adv2['allocation'], adv1['allocation']
                    advertiser_temp = copy.deepcopy(advertiser_can)
        X = Regret_new(advertiser_can)
        Y = Regret_new(advertiser_best)
        if X < Y and (Y-X) > 10 :
            advertiser_best = copy.deepcopy(advertiser_can)  
        else:
            break
    return advertiser_best
```

```python
final_advertisers = AdvertiserDriven(advertisers)
```

```python
total_regret = excessive_regret + unsatisfied_regret
print(total_regret)
end2 = time.time()
print(f"Total runtime of the program is {end2 - begin2}")
```
